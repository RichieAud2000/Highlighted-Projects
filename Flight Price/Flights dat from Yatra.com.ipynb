```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(3)

# Opening the homepage of yatra.in/schedule
url = "https://www.yatra.com/flight-schedule/"
driver.get(url)

time.sleep(2)
```

```python
# scrape all destinations urls
destination_urls = []
start=0
end=1
for page in range(start,end):#for loop for scrapping all destinations in one page
    url=driver.find_elements(By.XPATH,'/html/body/div[3]/section/aside/section[2]/section/ul[1]/li[1]/a')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))                       

#Make empty lists
Airline = []   
Date_of_Journey = []
Source = []
Destination = []
Route = []
Dep_Time = []
Arrival_Time = []
Duration = [] 
Total_Stops = []
Additional_Info = []
Price = []

for url in destination_urls:
    driver.get(url)
    time.sleep(2)
    
    # Extracting Airline from xpath
    try:
        Airline = driver.find_elements(By.XPATH,'//span[@class="i-b text ellipsis"]//span')      
        Airline.append(Airline.text)
    except NoSuchElementException:
        Airline.append('-')    


    time.sleep(1)
    # Extracting Date_of_Journey 
    try:
        Date_of_Journey = driver.find_elements(By.XPATH,'//div[@class="fs-12"]//p')      
        Date_of_Journey.append(Date_of_Journey.text)
    except NoSuchElementException:
        Date_of_Journey.append('-')


    time.sleep(1)    
    # Extracting Source from xpath    
    try:
        Source = driver.find_elements(By.XPATH,'//div[@class="fs-10 font-lightgrey no-wrap city ellipsis"]//p')  
        Source.append(Source.text)
    except NoSuchElementException:
        Source.append('-')


    time.sleep(1)
    # Extracting Destination from xpath
    try:
        Destination = driver.find_elements(By.XPATH,'//div[@class="fs-10 font-lightgrey no-wrap city ellipsis""]/p')  
        Destination.append(Destination.text)
    except NoSuchElementException:
        Destination.append('-')


    time.sleep(1)
    # Extracting Route from xpath    
    try:
        Route = driver.find_elements(By.XPATH,'//div[@class="mb-2 fs-13 no-wrap ellipsis"]/p')  
        Route.append(Route.text)
    except NoSuchElementException:
        Route.append('-')


    time.sleep(1)
    # Extracting Dep_Time from xpath
    try:
        Dep_Time = driver.find_elements(By.XPATH,'//div[@class="bold fs-16 mb-2"]')  
        Dep_Time.append(Dep_Time.text)
    except NoSuchElementException:
        Dep_Time.append('-')


    time.sleep(1)
    # Extracting Arrival_Time from xpath
    try:
        Arrival_Time= driver.find_elements(By.XPATH,'//div[@class="bold fs-16 mb-2"]/p')  
        Arrival_Time.append(Arrival_Time.text)
    except NoSuchElementException:
         Arrival_Time.append('-')


    time.sleep(1)
    # Extracting Duration from xpath         
    try:
        Duration= driver.find_elements(By.XPATH,'//span[@class="fs-12 du text-left mb-14"]')  
        Duration.append(Duration.text)
    except NoSuchElementException:
        Duration.append('-')



    time.sleep(1)
    # Extracting Total_Stops from xpath    
    try:
        Total_Stops= driver.find_element(By.XPATH,'//div[@class="fs-12 du mt-10 no-wrap"]/span')  
        Total_Stops.append(Total_Stops.text)
    except NoSuchElementException:
        Total_Stops.append('-')

    time.sleep(1)
    # Extracting Additional_Info from xpath
    try:
        Additional_Info = driver.find_elements(By.XPATH,'//div[@class="contents"]')  
        Additional_Info.append(Additional_Info.text)
    except NoSuchElementException:
        Additional_Info.append('No info')
        
    time.sleep(1)
    # Extracting Price from xpath
    try:
        Price = driver.find_elements(By.XPATH,'//div[@class="fs-15 mr-2 ytfi-rupee"]/i')  
        Price.append(Price.text)
    except NoSuchElementException:
        Price.append('-')
time.sleep(2)
# Make dataframe
df = pd.DataFrame({'Airline':Airline,'Date_of_Journey':Date_of_Journey,'Source':Source,'Destination':Destination,'Route':Route,
                        'Dep_Time':Dep_Time,'Arrival_Time':Arrival_Time,'Duration':Duration,
                        'Other Total_Stops':Total_Stops,'Additional_Info':Additional_Info,'Price':Price})
 
```

```python
df.to_excel(r'Flight_Data.xlsx', index=False)
```
