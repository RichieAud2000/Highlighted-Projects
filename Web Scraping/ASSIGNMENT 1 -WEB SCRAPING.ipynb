# IMDB TOP 100 Indian Movies

### Import Required Libraries

```python
from bs4 import BeautifulSoup
import requests


page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')#loading the page url for 1-100
page


soup = BeautifulSoup(page.content)
```

```python
page
```

```python
soup
```

# Scraping Multiple Titles

```python
titles= [] #empty list to store the titles

for i in soup.find_all('td',class_="titleColumn"):
    titles.append(i.text.split('\n')[2].strip())
    
titles=titles[:100]
titles
```

```python
year =[]  

for i in soup.find_all('td',class_="titleColumn"):
    year.append(i.text.split('\n')[3][1:-1])
    
year=year[:100]
year
```

```python
ratings = [] #empty list to store the titles

for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    ratings.append(i.text.split('\n')[1])
    
ratings=ratings[:100]
ratings
```

# Making a Dataframe

```python
import pandas as pd
df = pd.DataFrame({'Titles':titles,'Ratings':ratings,'Year Of Release':year})
df
```

# CNBC WORLD HEADLINES SCRAPE

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
```

```python
soup = BeautifulSoup(page.content) #Getting page content
```

```python
#now we have all the tags in which there are the elements data and titles

#now we will extract the text from theese tags one by one by looping over these tags

titles= [] #empty list to store the titles

for i in soup.find_all('div',class_="RiverPlusCard-container"):
    titles.append(i.get_text(strip=True,separator=" ").split('\n'))
    
titles
```

```python
def scrape():
    url = 'https://www.cnbc.com/world/?region=world'
    response = requests.get(url).content
    soup = BeautifulSoup(response, 'html.parser')
    areas = soup.find_all('div', class_='RiverPlusCard-container')
    for area in areas:
        print(area.find('a').attrs['href'])


if __name__ == '__main__':
    scrape()
```

# scrape mentioned details from dineout.co.in

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
soup = BeautifulSoup(page.content)
```

```python
#now we have all the tags in which there are the elements data and titles

#now we will extract the text from theese tags one by one by looping over these tags

titles= [] #empty list to store the titles

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
    
titles
```

```python
cuisines= [] #empty list to store the titles

for i in soup.find_all('span',class_='double-line-ellipsis'):
    cuisines.append(i.text.split('|')[1])
    
cuisines
```

```python
locations= [] #empty list to store the titles

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    locations.append(i.text)
    
locations 
```

```python
prices= [] #empty list to store the titles

for i in soup.find_all('span',class_='double-line-ellipsis'):
    prices.append(i.text.split('|')[0])
    
prices
```

```python
ratings= [] #empty list to store the titles

for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
    
ratings
```

```python
images= [] #empty list to store the link

for i in soup.find_all('img',class_='no-img'):
    images.append(i.get('data-src')) #and href will scrap the link of the objects or the pages
    
images
```
