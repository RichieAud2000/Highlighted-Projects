```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC # expected_conditions module contains a set of predefined conditions to use with WebDriverWait used for the visibility_of_element_located to be able to get the elements in the table as shown on website list  
import pandas as pd
import html5lib
import lxml
from bs4 import BeautifulSoup
```

```python
#To scrap data from multiple pages with the recruiters details , found 6120 posts and all are scrapped 

name =[]
designation=[]
company = []
skills = []
location = []

start = 0

end = 120

for page in range(start,end):
    try:
        a1 = driver.find_elements(By.XPATH,"//span[@class='fl ellipsis']")
        a2 = driver.find_elements(By.XPATH,'//span[@class="ellipsis clr"]')
        a3 = driver.find_elements(By.XPATH,'//li[@class="cmpIcn"]')
        a4 = driver.find_elements(By.XPATH,'//div[@class="hireSec highlightable"]')
        a5 = driver.find_elements(By.XPATH,'//li[@class="locIcn"]')
        wait_button = WebDriverWait(driver, 10)
        for i in a1:
            name.append(i.text)
        for i in a2:
            designation.append(i.text)
        for i in a3:
            company.append(i.text)
        for i in a4:
            skills.append(i.text)        
        for i in a5:
            location.append(i.text)
        next_button = driver.find_element(By.XPATH,'//button[@class="pagination"]') #to scrap data from the next page 
        next_button.click()
        time.sleep(3)
    except NoSuchElementException :  
            name.append('-')
            designation.append('-')
            company.append('-')
            skills.append('-')
            location.append('-')
```
